```python
# Label encoding, can do the same by using  LabelEncoder from scikitlearn
import pandas as pd
from sklearn import preprocessing
#read data
df = pd.read_csv("/home/vbdirtx5000/Desktop/thomtt/Approach/variables/train.csv")
#fill NaN value ở cột ord_2
#sử dụng fillna từ pandas vì LabelEncoder của scikit-learn hông giải quyết giá trị NaN
df.loc[:, "ord_2"]= df.ord_2.fillna("NONE")
# initialize LabelEncoder
lbl_enc = preprocessing.LabelEncoder()

#fit label encoder and transform values on ord_2 column,
df.loc[:, "ord_2"] = lbl_enc.fit_transform(df.ord_2.values)
```

```python

# check memory requirement 
import numpy as np
#create example feature matrix
example = np.array([
    [0,0,1],
    [1,0,0],
    [1,0,1]
])
print(example.nbytes)

print("------")
#we are only interested in 1s.
# any numpy array can be converted to a sparse matrix
import numpy as np
from scipy import sparse 
## convert numpy arr to sparse CSR matrix
sparse_example = sparse.csr_matrix(example)
#print size of this sparse matrix
print(sparse_example.data.nbytes)
```

```python
# concatenate of training set and test sets to learn about the categories present in the test set 
# but not in the training set or rare categories in the training set
import pandas as pd
from sklearn import preprocessing
#read training data
train = pd.read_csv("/home/vbdirtx5000/Desktop/thomtt/Approach/variables/train.csv")
#read test data
test = pd.read_csv("/home/vbdirtx5000/Desktop/thomtt/Approach/variables/test.csv")
#create a fake target column for test data, since this column doesn't exist
test.loc[:, "target"] = -1
#concatenate both training and test data
data = pd.concat([train, test]).reset_index(drop= True)

#make a list of features mà ta quan tâm, trong đó, id và target ko encode
features = [x for x in train.columns if x not in ["id", "target"]]

#loop over the features list
for feat in features:
    #tạo một instance của LabelEncoder cho mỗi feature
    lbl_enc = preprocessing.LabelEncoder()
    #vì categorical data, fillna với 1 string và convert tất cả data về dạng string
    # dù là int hay float thì cũng convert về string. 
    temp_col = data[feat].fillna("NONE").astype(str).values
    #sử dụng fit_transform, 
    data.loc[:, feat] = lbl_enc.fit_transform(temp_col)

#split train and test data again
train = data[data.target != -1].reset_index(drop = True)
test = data[data.target == -1].reset_index(drop= True)
```
